## 2. Now write a function that returns the largest element on a list

```python
def greatest(arr):
    maximo = 0
    while True:
        for nums in arr:       
            if nums > maximo:
                maximo = nums
            
        return maximo
    
```

## 6. Write a function that returns the factorial of a number.

```python
def factorial(n):
   
    factorial = 1
    
    for fact in range(1, n + 1):
        factorial *= fact
    
    return factorial
```

```python
# This will test your function 
test_factorial(factorial)
```
